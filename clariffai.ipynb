```python
from clarifai.rest import ClarifaiApp
from clarifai.rest import Workflow
from clarifai.rest import Image as ClImage
import numpy
import os
app = ClarifaiApp(api_key='7a168bd3200444d0a458140ad6c84f02')
workflow = Workflow(app.api, workflow_id="Demographics")

features=[]
DIR='/home/ubuntu/ME-beautydatabase/images/male/asian/'
for file in os.listdir(DIR):
   if file.endswith('.jpg') or file.endswith('.jpeg') or file.endswith('.png'):

        image_path = os.path.join(DIR, file)
        image = ClImage(file_obj=open(image_path, 'rb'))
        result = workflow.predict([image])
        concept_age= result['results'][0]['outputs'][4]['data']['regions'][0]['data']['concepts']
        age_10=0
        age_20=0
        age_30=0
        age_40=0
        for conc in concept_age:
            if (conc['name']== '10-19'):
                age_10 = numpy.round(conc['value'],2)     
            if (conc['name']== '20-29'):
                age_20 = numpy.round(conc['value'],2) 
            if (conc['name']== '30-39'):
                age_30 = numpy.round(conc['value'],2)   
            if (conc['name']== '40-49'):
                age_40 = numpy.round(conc['value'],2)           

        concept_gender= result['results'][0]['outputs'][3]['data']['regions'][0]['data']['concepts']
        male=0
        female=0
        for conc in concept_gender:
            if (conc['name']== 'Masculine'):
                male = numpy.round(conc['value'],2)     
            if (conc['name']== 'Feminine'):
                female = numpy.round(conc['value'],2)     

        concept_race= result['results'][0]['outputs'][2]['data']['regions'][0]['data']['concepts']
        white=0
        mideastern=0
        hispanic=0
        indian=0
        eastasian=0
        southeasian=0
        black=0
        #print(concept_race)
        for conc in concept_race:
            if (conc['name']== 'White'):
                white = numpy.round(conc['value'],2)     
            if (conc['name']== 'Black'):
                black = numpy.round(conc['value'],2) 
            if (conc['name']== 'Middle Eastern'):
                mideastern = numpy.round(conc['value'],2)     
            if (conc['name']== 'Latino_Hispanic'):
                hispanic= numpy.round(conc['value'],2) 
            if (conc['name']== 'East Asian'):
                eastasian= numpy.round(conc['value'],2)     
            if (conc['name']== 'Southeast Asian'):
                southeasian= numpy.round(conc['value'],2)
            if (conc['name']== 'Indian'):
                indian= numpy.round(conc['value'],2)  

        features.append([image_path,male,female,age_10, age_20,age_30, age_40, white, black, mideastern, hispanic, indian, southeasian, eastasian])
import pandas as pd
df=pd.DataFrame(features, columns=['path','male','female','age_10', 'age_20','age_30', 'age_40', 'white', 'black', 'mideastern', 'hispanic', 'indian', 'southeasian', 'eastasian'])
df.to_excel(clarifai_male_asian.xlsx') 
```

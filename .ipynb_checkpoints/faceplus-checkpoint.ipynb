```python
#betaface get data

import requests
import base64
import xml.etree.ElementTree as ET
import xmltodict
import json
import time

import os
import pandas as pd

# https://www.betafaceapi.com/wpa/index.php/documentation
# https://mgng.mugbum.info/679

API_KEY = "59208710-4beb-11ea-9287-0cc47a6c4dbd"
API_SECRET = "592826c7-4beb-11ea-9287-0cc47a6c4dbd"

def image_encoder(image_file_path):
    """
        画像データをbase64形式にエンコードする
    """
    with open(image_file_path, "rb") as f:
        data = f.read()

    return base64.b64encode(data)

def get_img_uid(image_file_path):
    """
        img_uidを取得する関数
    """
    img_data = image_encoder(image_file_path)

    param = {
        "api_key" : API_KEY,
        "api_secret" : API_SECRET,
        #"detection_flags" : ,
        "imagefile_data" : img_data,
        "original_filename" : "face.jpg"
    }

    res = requests.post("http://betafaceapi.com/service.svc/UploadNewImage_File", data=param)
    dict = xmltodict.parse(res.content)
    #print json.dumps(dict, indent=1)

    img_uid = dict["BetafaceImageResponse"]["img_uid"]
    
    return img_uid

def get_img_info(img_uid,file):
    """
        顔の特徴を辞書型で返す
    """
    #img_data = image_encoder(file)
    
    param = {
        "api_key" : API_KEY,
        "api_secret" : API_SECRET,
        "img_uid" : img_uid
    }

    res = requests.post("http://betafaceapi.com/service.svc/GetImageInfo", data=param)
   # time.sleep(5)
    
    dict = xmltodict.parse(res.content)
    face_info = dict["BetafaceImageInfoResponse"]["faces"]["FaceInfo"]['tags']
    df = pd.DataFrame(face_info, columns=face_info.keys())
    

    params_dict = {}
    
    params_dict['path']=file
    
    for index, row in df.iterrows():
   
        line =df.iloc[index].get(key='TagInfo')
        
       # if (params_dict[line['name']] == 'age') or (params_dict[line['name']] == 'gender') or (params_dict[line['name']] == 'expression') or (params_dict[line['name']] == 'race'):
         
        if (line['name'] == 'age') or (line['name'] == 'gender') or (line['name'] == 'expression') or (line['name'] == 'race'):
                
               params_dict[line['name']]= str(line['value'])+','+str(line['confidence'])
           
        else:
            if line['value']=='yes':
                value = line['confidence']
            else: 
                value=-float(line['confidence'])
            params_dict[line['name']]=value  
        

        
        
       # print(params_dict)
        
        
        
    return params_dict

DIR='/home/ubuntu/ME-beautydatabase/images/female/indian/'
#DIR='/home/ubuntu/shlak/'
#DIR='/home/ubuntu/face_images2/'
faces=[]
import cv2
code={}
for file in os.listdir(DIR):
   if file.endswith('.jpg') or file.endswith('.jpeg') or file.endswith('.png'):
       path= DIR + file
       image_file_path = path
       
       img_uid = get_img_uid(image_file_path)
       code[path]=img_uid
       #img_info = get_img_info(img_uid,image_file_path)
       #faces.append(img_info)

for file in os.listdir(DIR):
   if file.endswith('.jpg') or file.endswith('.jpeg') or file.endswith('.png'):
       path= DIR + file
       image_file_path = path
       
       #img_uid = get_img_uid(image_file_path)
       #code[path]=img_uid
       print(image_file_path)
       print(code[path])
       try: 
           img_info = get_img_info(code[path],image_file_path)
           faces.append(img_info)
       except:
           print('nonono')
df=pd.DataFrame(faces)
        
df.to_excel('betaface_female_indian.xlsx')
```
